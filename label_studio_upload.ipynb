```python
def create_task(indices_eos_predicted: list, file_text: str, tasks: list) -> None:
    indices_eos_predicted.sort()
    sents = []

    i = 0
    for index in indices_eos_predicted:
        if(i == 0):
            sents.append((0, index))
        else:
            sents.append((indices_eos_predicted[i - 1] + 1, index))
        i += 1

    sents.append((indices_eos_predicted[-1] + 1, len(file_text)))
    
    task = {}
    task['data'] = {}
    task['data']['text'] = file_text

    task['predictions'] = []
    prediction = {}
    prediction['result'] = []

    for start, end in sents:
        prediction['result'].append({
            'from_name' : 'text_labels',
            'to_name' : 'text_object',
            'type' : 'labels',
            'value' : {
                'start' : start,
                'end' : end,
                'text' : file_text[start:end],
                'labels' : ['Sentence']
            }
        })


    task['predictions'].append(prediction)
    tasks.append(task)
```

```python
file_dir = './FINAL/files'
file_index_dir = './FINAL/index_files'

tasks = []
cnt = 0
def create_tasks(file_dir : str, file_index_dir : str):
    # global cnt
    global tasks
    for fname in os.listdir(file_dir):
        # print(fname)
        for f_indexname in os.listdir(file_index_dir):
            if('indexfile_of_' + fname == f_indexname + '.txt'):
                print('Creating task out of ' + fname)
                indices_eos_predicted = []
                eos_predicted = open(file_index_dir + '/' + f_indexname, 'r') 
                indices_eos_predicted = list(map(int, eos_predicted.read().split(',')))
                text_file = open(file_dir + '/' + fname, 'r')
                file_text = text_file.read()

                create_task(indices_eos_predicted, file_text, tasks)

                print('Created task from ' + fname + 'successfully.')
                print('---'*10)

create_tasks(file_dir=file_dir, file_index_dir=file_index_dir)
```

```python
project.import_tasks(
    tasks
)
```
